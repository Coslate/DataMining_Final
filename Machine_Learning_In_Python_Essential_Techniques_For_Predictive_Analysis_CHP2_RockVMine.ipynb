# Way1 - Use urlopen to get data

```python
from   urllib.request import urlopen
import sys

target_url = ("https://archive.ics.uci.edu/ml/machine-learning-databases/undocumented/connectionist-bench/sonar/sonar.all-data")
data_urlopen = urlopen(target_url)
data_urlopen
```

```python
X = []
Y = []

for line in data_urlopen:
    #row0 = line
    #print(f"row0 = {row0}")
    #row1 = line.strip()
    #print(f'row1 = {row1}')
    row = bytes.decode(line.strip())
    #print(f"row2 = {row}")
    row = row.split(',')
    #print(f'row[0] = {row[0]}')
    X.append(row)

#print(X)
print('Number of Rows of Data = {x}'.format(x = len(X)))
print('Number of Columns of Data = {y}'.format(y = len(X[1])))
```

# Way2 - Use BeautifulSoup to get data

```python
import requests
from   bs4 import BeautifulSoup

#Disable Warning
requests.packages.urllib3.disable_warnings()

#Intrepretining the response
response = requests.get(target_url, cookies = {'over18':"1"}, verify = False)
soup     = BeautifulSoup(response.text, 'lxml')
print(soup.prettify())
```

```python
import re

X = []
Y = []

data_core = soup.find('p').get_text()
data_core_list = data_core.split('\n')

for data_line in data_core_list:
    if(re.match(r'\S+', data_line)):
        #print(f'row0 = {data_line}')
        row1 = data_line.strip()
        #print(f'row1 = {row1}')
        row = data_line.strip().split(',')
        X.append(row)
    
print('Number of Rows of Data = {x}'.format(x = len(X)))
print('Number of Columns of Data = {y}'.format(y = len(X[0])))
```

```python
import pandas as pd

header_col = ['Attr'+str(x) for x in range(0, (len(X[0])-1))]
header_col.append('Label')
df = pd.DataFrame(X, columns=header_col)
df
```

#### Calculate the Pearson Correlation

```python
import math

def CalculateCorr(data_row0, data_row1, org_data_frame):
    row_num = len(org_data_frame.values)
    col_num = len(org_data_frame.columns)    
    mean_data_row0 = 0
    mean_data_row1 = 0
    var0_sum = 0
    var1_sum = 0
    corr_coeff = 0
    
    for i in range(row_num):
        mean_data_row0 += data_row0[i]
        mean_data_row1 += data_row1[i]
    mean_data_row0 /= row_num
    mean_data_row1 /= row_num
    
    for i in range(row_num):
        var0_sum += math.pow((data_row0[i] - mean_data_row0), 2)
        var1_sum += math.pow((data_row1[i] - mean_data_row1), 2)

    for i in range(row_num):
        corr_coeff += (data_row0[i] - mean_data_row0)*(data_row1[i] - mean_data_row1)

    corr_coeff /= math.sqrt(var0_sum*var1_sum)
    
    return corr_coeff
```

```python
data_row2 = rock_mines_df.iloc[0:row_num, 1]
data_row3 = rock_mines_df.iloc[0:row_num, 2]
data_row21 = rock_mines_df.iloc[0:row_num, 20]

corr23 = CalculateCorr(data_row2, data_row3, rock_mines_df)
print(f'corr23 = {corr23}')
corr221 = CalculateCorr(data_row2, data_row21, rock_mines_df)
print(f'corr221 = {corr221}')
```
