```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pprint
import time
from sklearn.cluster import KMeans
from pylab import rcParams
%matplotlib inline

team = pd.read_excel("./NBA_revised/team_playoff_all.xlsx")
team
```

```python
team_df = team.groupby('Team').mean()
print(f"len(team_df) = {len(team_df)}")
team_df.head(len(team_df))
```

```python
total_team = list(team['Team'])
team_set = set().union(total_team)
team_set
```

```python
def TotalWin(team_name):
    return len(team[(team.Team == team_name) & (team['W/L'] == 'W')])
def TotalLose(team_name):
    return len(team[(team.Team == team_name) & (team['W/L'] == 'L')])
def TotalGame(team_name):
    return len(team[(team.Team == team_name)])

win_hash = {i: TotalWin(i) for i in team_set}
lose_hash = {i: TotalLose(i) for i in team_set}
total_game_hash = {i: TotalGame(i) for i in team_set}
total_game_hash
```

```python
l = list(team['Team'])
l_win = [win_hash[x] for x in l]
l_lose = [lose_hash[x] for x in l]
l_total = [total_game_hash[x] for x in l]
l_win_percent = [win_hash[x]/total_game_hash[x] for x in l]

team['Win_Game'] = l_win
team['Lose_Game'] = l_lose
team['Total_Game'] = l_total
team['Win_Percentage'] = l_win_percent
team_df = team.groupby('Team').mean()
team_df.head(len(team_df))
```
