```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def ConvertToYX(team_file, oppo_file):
    team = pd.read_csv(team_file)
    oppo = pd.read_csv(oppo_file)
    
    team_revised_x = team.drop(columns = ['Team', 'Date', 'Score', 'W/L', 'Opponent', 'PTS'])
    oppo_revised_x = oppo.drop(columns = ['Team', 'Date', 'Score', 'W/L', 'Opponent', 'PTS'])
    
    team_label_list = team['W/L']
    team_label_list = list(team_label_list)
    column_team_revised_x = list(team_revised_x.columns)
    new_data_frame_columns_left = [x for x in column_team_revised_x]
    
    column_oppo_revised_x = list(oppo_revised_x.columns)
    new_data_frame_columns_right = ['oppo_'+x for x in column_oppo_revised_x]
    
    new_data_frame_columns = new_data_frame_columns_left + new_data_frame_columns_right + ['Label']
    new_data_frame = pd.DataFrame(columns=new_data_frame_columns)
    
    for row_id, row in enumerate(team_revised_x.values):
        oppo_row = list(oppo_revised_x.iloc[row_id])
        oppo_row.append(team_label_list[row_id])
        new_data_frame.loc[row_id] = list(row)+list(oppo_row)
    
    #For Visual
    new_data_frame.to_csv('./nba_extracted_XY.csv', encoding='utf-8', index=False)
    
    Y = new_data_frame['Label']
    X = new_data_frame.drop(columns = ['Label'])
    
    return(Y, X)
```

```python
team_file = "../NBA_Ino_part/nba_extracted_team.csv"
oppo_file = "../NBA_Ino_part/nba_extracted_oppo.csv"
(Y, X) = ConvertToYX(team_file, oppo_file)
```
